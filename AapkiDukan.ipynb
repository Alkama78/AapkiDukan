```python
print("Welcome to AapkiDukan")
customer_name = input("Enter your good name, please.\n ")
print()
print(f"Welcome, {customer_name}! How are you doing?")
print("What would you like to have?")
print("Please check our Menu card.")

menu = {1: 'Burger', 2: 'Fries', 3: 'Coke', 4: 'McFlurry'}
prices = {'Burger': 50, 'Fries': 40, 'Coke': 40, 'McFlurry': 50}
print()
final_order = []

def show_menu():
    print()
    print("Menu available")
    for item_num, item_name in menu.items():
        print("{} {}".format(item_num, item_name))
    print()

    print("Price of items available")
    for item_name, price in prices.items():
        print("{} {}".format(item_name, "Rs." + str(price)))
    print()

def add_item():
    final_amount = 0
    try:
        item_num = int(input("Enter item number :- "))
    except ValueError:
        print("Invalid input. Please enter a number between (1-4).")
        return
    
    if item_num not in menu:
        print("Invalid input. Please enter a number between (1-4).")
        return

    try:
        no_of_order = int(input("Enter number of orders :- "))
    except ValueError:
        print("Invalid input. Please enter a valid number.")
        return

    item_name = menu[item_num]
    total_price = prices[item_name] * no_of_order
    final_order.append(f"{item_name} x {no_of_order} = Rs. {total_price}")
    print()
    print(f'Great choice! {no_of_order} {item_name} added to your cart.')
    print()

def show_cart():
    if not final_order:
        print("Your cart is empty.")
    else:
        print("Items in your cart:")
        for item in final_order:
            print(item)
    print()

def show_bill():
    if not final_order:
        print("Sorry as you don't have any orders.\nWe can't provide you the bill")
        print('Would you like to order something?\nSo that we can able to provide bill to you.')
        user_confirmation=int(input('If Yes(1) / NO(0) :-'))
        if user_confirmation == 0:
            exit()
        else:
            show_menu()
            add_item()
        print()
        
    else:
        total_amount = sum(prices[item.split(' x ')[0]] * int(item.split(' x ')[1].split(' = ')[0]) for item in final_order)
        print('''|~~~~~~~~~~~~~~~Your Bill~~~~~~~~~~~~~~|''')
        print('''|--------**********************--------|''')
        print('''| Cannaught Plaza Restaurants Pvt. Ltd.|\n| AapkiDukan 77-88098-980              |\n| Service Tax No. :- 077ACCOYP9873G    |\n|--------*********************---------|''')
        print("| 6A-6B PVR Complex, Vikas Puri,       |")
        print("| New Delhi - 18                       |")
        print("| Phone No. 8978675698                 |")
        print("| TIN - 088997788                      |")
        print('''|--------*********************---------|\n|| Name   || Price || Qty || Amount   ||\n|--------*********************---------|''')
        for item in final_order:
            print(item)
        print(f"Total Amount: Rs. {total_amount}")
        print('''|--------*********************---------|''')
        print('''|--------*********************---------|''')
        print('''| Thanks for coming. We hope you loved |\n| our services\t\t\t\t|''')
        print('''|~~~~~~~~~~~~~~~Your Bill~~~~~~~~~~~~~~|''')
        print()

        print("If you proceed with the generation of bill then you won't be able to add item in the same cart again.\n")
        print("Do want to generate bill ?")
        
        user_confirmation = int(input("If Yes(1) / No(0) :- "))
        print()
        if user_confirmation == 1:
            total_amount = sum(prices[item.split(' x ')[0]] * int(item.split(' x ')[1].split(' = ')[0]) for item in final_order)
            print('''|~~~~~~~~~~~~~~~Your Bill~~~~~~~~~~~~~~|''')
            print('''|--------**********************--------|''')
            print('''| Cannaught Plaza Restaurants Pvt. Ltd.|\n| AapkiDukan 77-88098-980              |\n| Service Tax No. :- 077ACCOYP9873G    |\n|--------*********************---------|''')
            print("| 6A-6B PVR Complex, Vikas Puri,       |")
            print("| New Delhi - 18                       |")
            print("| Phone No. 8978675698                 |")
            print("| TIN - 088997788                      |")
            print('''|--------*********************---------|\n|| Name   || Price || Qty || Amount   ||\n|--------*********************---------|''')
            for item in final_order:
                print(item)
            print(f"Total Amount: Rs. {total_amount}")
            print('''|--------*********************---------|''')
            print('''|--------*********************---------|''')
            print('''| Thanks for coming. We hope you loved |\n| our services\t\t\t\t|''')
            print('''|~~~~~~~~~~~~~~~Your Bill~~~~~~~~~~~~~~|''')
            final_order.clear()
        else:
            show_menu()
            add_item()
    print()

def delete_order():
    final_order.clear()
    print("All items deleted from cart.")
    print()

def payment():
    print("Payment Methods:")
    print("1. Cash")
    print("2. Credit Card")
    print("3. Debit Card")
    print("4. UPI")
    print("5. Wallet")
    print()
    choice = input("Please select a payment method: ")
    print(f"Payment successful via {choice}. Thank you!")
    print()

while True:
    print("1. Show Menu")
    print("2. Add items to cart")
    print("3. Show items in cart")
    print("4. Show bill")
    print("5. Delete all items from cart")
    print("6. Make Payment")
    print("7. Exit")
    print()

    try:
        choice = int(input("Enter your choice (1-7): "))
        if choice == 1:
            show_menu()
        elif choice == 2:
            add_item()
        elif choice == 3:
            show_cart()
        elif choice == 4:
            show_bill()
        elif choice == 5:
            delete_order()
        elif choice == 6:
            payment()
        elif choice == 7:
            if not final_order:
                print('''You didn't order anything.\nWe feel bad that you didn't find your product.\nPlease give us feedback so that we can implement our services.''')
                print("Please suggest a product which you had not found in our platform.")
                print()
                product_suggestion=input("Enter the Product name please :- ")
                print('Thanks for your prompt response.')
                print()
                print(f'Great Choice! {product_suggestion}')
                print('We will add this product in our menu as earliest as possible.')
                print(f"Thank you, {customer_name}! Have a great day!")
                break
            else:
                print("You have some items in the cart.")
                confirmation_from_user = int(input("Would you like to proceed with exit(0) or order(1)?  "))
                if confirmation_from_user==1:
                   continue
                else:
                    break
        else:
            print("Invalid choice. Please enter a number between 1 and 7.")
    except ValueError:
        print("Invalid input. Please enter a valid number.")

```
